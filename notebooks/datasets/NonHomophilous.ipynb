```python
from cool_graph.datasets import NonHomophilous
from cool_graph.runners import Runner, HypeRunner
from sklearn.model_selection import train_test_split
```

```python
root = './data/nonhomophilous'
```

# Genius

```python
genius = NonHomophilous(root=root, name="Genius")
```

```python
idx = list(range(genius.data.x.shape[0]))
train_idx, test_idx = train_test_split(idx, test_size=0.4, random_state=42)
valid_idx, test_idx = train_test_split(test_idx, test_size=0.5, random_state=42)
print(len(train_idx), len(valid_idx), len(test_idx))
```

```python
runner = HypeRunner(
    data=genius.data,
    seed=42,
    train_idx=train_idx,
    test_idx=valid_idx,
    metrics=['accuracy']
)
```

```python
run_result=runner.optimize_run(n_trials=50)
```

```python
result=runner.predict_proba(genius.data, test_idx, metrics=['accuracy'])
```

```python
run_result2=runner.optimize_run(n_trials=50)
```
