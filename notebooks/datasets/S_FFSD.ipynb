```python
from cool_graph.datasets import S_FFSD
from cool_graph.runners import Runner, HypeRunner
from sklearn.model_selection import train_test_split
```

```python
root = './data/s_ffsd'
```

# S-FFSD

```python
s_ffsd = S_FFSD(root=root)
```

```python
idx = list(range(s_ffsd.data.x.shape[0]))
train_idx, test_idx = train_test_split(idx, test_size=0.4, random_state=42)
valid_idx, test_idx = train_test_split(test_idx, test_size=0.5, random_state=42)
print(len(train_idx), len(valid_idx), len(test_idx))
```

```python
runner = HypeRunner(
    data=s_ffsd.data,
    seed=42,
    train_idx=train_idx,
    test_idx=valid_idx,
    metrics=['roc_auc']
)
```

```python
result=runner.optimize_run(n_trials=50)
```

```python
result = runner.predict_proba(s_ffsd.data, test_idx, metrics=['roc_auc'])
```
